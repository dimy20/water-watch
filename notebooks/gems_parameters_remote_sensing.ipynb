```python
pigments = pd.read_csv(dir / "Pigment.csv", encoding="ISO-8859-1")
pigments.head()
```

```python
# contar cuantos registros hay de este parametro en uruguay despues del 2015
def uy_param_count_after_2015(param_code, dataset, countries = ("URY")):
	param_only = dataset[dataset["Parameter Code"] == param_code]
	param_only_uy = param_only[param_only['GEMS Station Number'].str.startswith(countries, na=False)]
	param_only_uy_post_2015 = param_only_uy[pd.to_datetime(param_only_uy['Sample Date']).dt.year >= 2015]
	total_params = len(param_only_uy_post_2015)
	total_stations = len(param_only_uy_post_2015["GEMS Station Number"].unique())
	return total_params, total_stations
```

```python
cant_clorofila_a = uy_param_count_after_2015(param_code=chla_code, dataset=pigments)
cant_clorofila_a
```

```python
param_count = []
station_count = []

params_of_interest = [
	("TURB", "Optical.csv"),
	("TSS", "Water.csv"),
	("Chl-a", "Pigment.csv"),
	("TRANS", "Optical.csv"),
	#("Phyt-B", "Phytoplankton.csv")
]

for p in params_of_interest:
	param_code, fname = p[0], p[1]
	count = uy_param_count_after_2015(
		param_code,
		dataset=pd.read_csv(dir/fname, encoding="ISO-8859-1"),
		countries=("URY")
	)
	print(count)
	param_count.append(count[0])
	station_count.append(count[1])
	#print(count)
```

```python
print("Total parametros en Uruguay despues del 2015")

param_names = []

for p in params_of_interest:
	param_names.append(df[df["Parameter Code"] == p[0]]["Parameter Name"].item())


total_params = {
	"Code": [p[0] for p in params_of_interest] + ["Total"],
	"Parametro": param_names + [len(param_names)],
	"Cant": param_count + [sum(param_count)],
	"Cant Stations": station_count + [sum(station_count)]
}

total_df = pd.DataFrame(total_params)
total_df
```
