```python
# contar cuantos registros hay de este parametro en uruguay despues del 2015
def uy_param_count_after_2015(param_code, dataset, after=2015, countries = ("URY")):
	param_only = dataset[dataset["Parameter Code"] == param_code]
	param_only_uy = param_only[param_only['GEMS Station Number'].str.startswith(countries, na=False)]
	param_only_uy_post_2015 = param_only_uy[pd.to_datetime(param_only_uy['Sample Date']).dt.year >= after]
	total_params = len(param_only_uy_post_2015)
	stations = param_only_uy_post_2015["GEMS Station Number"].unique()
	#total_stations = len()
	return total_params, stations
```

```python
params_of_interest = [
    ("pH", ["pH.csv"]),
    ("ECOLI", ["Indicator_Organism.csv"]),
    ("FECALCOLI", ["Indicator_Organism.csv"]),
    ("TOTCOLI", ["Indicator_Organism.csv"]),
    ("TOTMESO", ["Indicator_Organism.csv"]),
    ("ENTCOCC", ["Indicator_Organism.csv"]),
    ("FECALSTREP", ["Indicator_Organism.csv"]),
    ("BOD", ["Oxygen_Demand.csv"]),
    ("COD", ["Oxygen_Demand.csv"]),
    ("PV", ["Oxygen_Demand.csv"]),
    ("O2-Dis", ["Dissolved_Gas.csv"]),
    ("O2-Dis-Sat", ["Dissolved_Gas.csv"]),
    ("NH3N", ["Other_Nitrogen.csv", "Oxidized_Nitrogen.csv"]),
    ("NH4N", ["Other_Nitrogen.csv", "Oxidized_Nitrogen.csv"]),
	("TDP", ["Phosphorus.csv"]),
	("TP", ["Phosphorus.csv"]),
	("TPP", ["Phosphorus.csv"]),
    ("PN", ["Other_Nitrogen.csv", "Oxidized_Nitrogen.csv"]),
    ("TN", ["Other_Nitrogen.csv", "Oxidized_Nitrogen.csv"]),
    ("TDN", ["Other_Nitrogen.csv", "Oxidized_Nitrogen.csv"]),
	("TEMP", ["Temperature.csv"]),
	("EC", ["Electrical_Conductance.csv"])
]
```

```python
param_counts = []
station_counts= []
stay_params = []
total_unique_stations = set()

for p in params_of_interest:
	param_code, files= p[0], p[1]

	param_count = 0
	station_count = 0

	for fname in files:
		count, stations = uy_param_count_after_2015(
			param_code,
			dataset=pd.read_csv(dir/fname, encoding="ISO-8859-1"),
			after=2015,
			countries=("URY")
		)

		param_count += count
		station_count += len(stations)

		for station in list(stations):
			total_unique_stations.add(station)

	if param_count > 0:
		param_counts.append(param_count)
		station_counts.append(station_count)
		stay_params.append(param_code)
```

```python
print("Total parametros en Uruguay despues del 2015")

param_names = []

for p in stay_params:
	param_names.append(df[df["Parameter Code"] == p]["Parameter Name"].item())

total_params = {
	"Code": [p for p in stay_params] + ["Total"],
	"Parametro": param_names + ["Total"],
	"Cantidad": param_counts + [sum(param_counts)],
	"Estaciones": station_counts + [len(total_unique_stations)]
}

total_df = pd.DataFrame(total_params)
total_df
```
