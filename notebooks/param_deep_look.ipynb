```python
import pandas as pd
from pathlib import Path
dir = Path("../GFQA_v3")
```

```python
def filtrar_uy_param(file, param_code):
	df = pd.read_csv(dir / file)
	param_only = df[df["Parameter Code"] == param_code]
	param_only_uy = param_only[param_only['GEMS Station Number'].str.startswith("URY", na=False)]
	return param_only_uy
```

```python
df_ph = filtrar_uy_param("pH.csv", "pH")
df_ph.head()
```
